# Matplotlib Notes

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Quick Plotting

### Line Plotting

```python
x = np.linspace(0, 15, 30)
y = np.sin(x) + 0.1*np.random.randn(len(x))
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, "o--", lw=0.7, ms=4)
```

```python
# Change the figure aspect ratio
plt.figure(figsize=(7,3))
plt.plot(x, y, "o", color="purple")
```

### Axis Labels

```python
plt.plot(x, y)
plt.xlabel("Time [s]")
plt.ylabel("Voltage [V]")
```

### Add a legend

```python
plt.plot(x, y, label="Component 1")
plt.xlabel("Time [s]")
plt.ylabel("Voltage [V]")
plt.legend(loc="lower left", fontsize=10)
```

### Multiple Lines

```python
x2 = np.linspace(0, 15, 30)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(7, 3))
plt.plot(x, y, "o", lw=0.7, ms=4, label="Component 1")
plt.plot(x2, y2, label="Component 2")
plt.xlabel("Time [s]")
plt.ylabel("Voltage [V]")
plt.legend(loc="lower left", fontsize=8)
```

## Histograms

### Quick Histograms

```python
res = np.random.randn(1000) * 0.2 + 0.4
```

```python
plt.hist(res)
```

### Change no. of bins

```python
plt.hist(res, bins=30, density=True)
plt.show()
```

### Multiple plots

```python
res2 = np.random.randn(1000) * 0.2 + 0.4
```

```python
plt.figure(figsize=(7, 3))
plt.hist(res, bins=30, density=True, histtype="step")
plt.hist(res2, bins=30, density=True, histtype="step")
plt.show()
```

## A More Advanced API

For more complicated plots such as:
* multiple subplots on 1 fig
* conveniently adding text to plots
* creating animations

One needs to use a slightly more advanced API

### Multiple Axes in 1 Figure

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
ax = axes[0][0]
ax.plot(x, y)
ax.set_xlabel("x label")
ax.set_ylabel("y label")

ax = axes[0][1]
ax.hist(res)
ax.set_xlabel("other x label")
ax.set_ylabel("other y label")

# Add text using .text() method
ax.text(0.1, 0.4, "sample text", transform=ax.transAxes)

# Change size of ax values
ax.tick_params(axis="both", labelsize=5)
```
